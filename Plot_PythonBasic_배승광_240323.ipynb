```python
# 그래프와 시각화
#matplotlib을 사용하기 위한 기본 설정
import numpy as np
import pandas as pd
PREVIOUS_MAX_ROWS = pd.options.display.max_rows
pd.options.display.max_rows = 20
np.random.seed(12345)

import matplotlib.pyplot as plt
import matplotlib
plt.rc('figure', figsize=(10, 6))
np.set_printoptions(precision=4, suppress=True)
```

```python
#matplotlib에서 그래프를 그리는 라이브러리 : pyplot
#plt라는 약어로 많이 사용됨
import matplotlib.pyplot as plt
```

```python
#그래프를 그리기 위한 데이터를 생성한다
#0-9 사이의 정수를 담은 배열을 생성한다
import numpy as np
data = np.arange(10)
print('데이터:', data)
print('-'*100)
#그래프를 그리기 위해서는 pyplot 라이브러리의 plot 함수를 사용하면 된다.
#함수의 인자로는 그래프로 표현할 데이터를 넣어준다
#plot 함수는 데이터만 인자로 넘어오는 경우 x축을 자동설정한다. 데이터는 y축에 사용되었다
plt.plot(data)
```

```python
#만약 x축, y축을 수동설정하고 싶다면 아래와 같이 하면 된다
xs = np.arange(0, 10)
ys = 2**xs
print('x축: ', xs)
print('y축: ', ys)
print('-'*100)
plt.plot(xs,ys)
plt.plot(ys)
```

```python
# 1.1.1 Figures and Subplots
# matplotlib에서 그래프는 Figure 객체 내에 존재함
# 생성된 figure는 비어있음 -> 여기에 그래프를 추가하면 된다.
fig = plt.figure()
```

```python
#figure에 4개의 그래프를 그리기
#subplot은 figure안에 들어가는 작은 그래프를 의미함
#add_subplot 3개의 인자를 받음 (행, 열, 현재 추가하는 그래프의 번호) 순서
ax1 = fig.add_subplot(2, 2, 1) # 2행 2열 크기의 subplot판 -> 1번째에 subplot을 추가
```

```python
#subplot을 2개 더 추가한다. 이번에는 2행 2열 크기의 figure에 2번째, 3번째 칸에 그래프를 추가한다
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

```python
fig
```

```python
#가장 마지막에 추가한 3번 subplot에 그래프 그리기
#랜덤한 수를 50개 담은 1차원 배열 생성
xs = np.random.randn(50)
print('xs: ', xs)
print('-'*100)

ax3.plot(xs,'k--') #3번째 subplot 객체는 ax3,
#'k--'는 검은색 점선으로 그리기위한 옵션
# 'b*' 옵션을 주면 파란색 별표로 그래프가 그려짐

fig # subplot이 그려졌는지 확인하기 위해 전체 그림인 figure를 확인
```

```python
# 1,2번 subplot에도 데이터 넣어 그래프 그리기

# 1번 그래프에는 히스토그램 넣기
# 3번 그래프를 그린 것처럼 subplot 객체의 메서드를 활용. hist는 히스토그램
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)

# 2번 그래프에는 산포도 그래프를 그림
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
fig
```

```python
# 1.1.2 subplot 간의 간격 조절하기
# matplotlib은 subplot 간에 적당한 간격(spacing)과 여백(padding)을 추가해준다
# 이 간격을 직접 조절해보기
# subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)

# 2행 2열 크기의 figure에 4개의 subplot에 히스토그램을 가지고 있다.
# x축, y축 subplot이 서로 겹쳐있음
# -> 이런 경우 subplot간의 간격을 조절해야함

fig, axes = plt.subplots(2, 2)
for i in range(2):
  for j in range(2):
    axes[i,j].hist(np.random.randn(5000), bins=50, color='k', alpha=0.5)
    axes[i,j].set_xlabel('xs')
    axes[i,j].set_ylabel('ys')
    axes[i,j].set_title('subplot overlap')
```

```python
#같은 그래프를 먼저 그린후에 subplot간의 간격을 조절해본다
fig, axes = plt.subplots(2, 2)
for i in range(2):
  for j in range(2):
    axes[i, j].hist(np.random.randn(5000), bins=50, color='k', alpha=0.5)
    axes[i,j].set_xlabel('xs')
    axes[i,j].set_ylabel('ys')
    axes[i,j].set_title('subplot overlap')

# subplot간의 간격을 조절
# wspace(width space) : fig의 40% 정도로 조절, hspace(height space) : fig의 60%정도로 조절하기
plt.subplots_adjust(wspace=0.4, hspace=0.6)
```

```python
#예를 들어 녹색 점선으로 그려진 x대 y그래프는 아래처럼 나타낼 수 있다
xs = [1,2,3,4,5]
ys = [1,2,3,4,5]
plt.plot(xs,ys,'g--')
```

```python
# linestype, color 옵션 설정
xs = [1,2,3,4,5]
ys = [1,2,3,4,5]
plt.plot(xs,ys, linestyle='--', color='g')
```

```python
# 특정 지점의 실제 데이터를 돋보이게 하기 위해 마커를 추가
# 마커는 색상 Character 다음에 마커 Character가 오고 그 다음에 선 스타일 Character가 온다
from numpy.random import randn
#검은색 선, 마커는 0, 선 종류는 점선(--)
plt.plot(randn(30).cumsum(),'ko--')
```

```python
#마커 역시도 옵션을 명시할 수 있다
plt.plot(randn(30).cumsum(), color='k', linestyle='dashed', marker='o')
```

```python
#아래는 같은 데이터를 사용하더라도 다른 옵션을 주어 다르게 보여줄 수 있다
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.plot(data, 'b^', label='triangle')

#범례를 보여준다
plt.legend(loc='best')
```

```python
# 1.1.4 눈금, 라벨, 범례
#x축 눈금이 포함된 간단한 그래프 2개
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(np.random.randn(1000).cumsum())

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(np.random.randn(1000).cumsum())
```

```python
#x축의 눈금을 변경하기 위한 가장 쉬운 방법은 set_xticks와 set_xticklabels 메서드를 사용하는 것이다
#set_xticks 메서드는 전체 데이터 범위를 따라 눈금을 어디에 배정할지 지정한다
#set_xticklabels 메서드는 다른 눈금 이름을 지정하고 싶을 때 사용한다
ticks = ax1.set_xticks([0, 250, 500, 750, 1000])
labels = ax2.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
fig
```

```python
#set_xlabel 메서드는 x축에 대한 이름을 지정한다
#set_title 메서드는 제목을 지정한다
ax1.set_title('My first matplotlib plot')
ax2.set_xlabel('Stages')
fig
```

```python
#다음과 같이 딕셔너리를 활용해 옵션을 한번에 주고 설정하는 것도 가능함
props = {
    'title':'My First matplotlib plot',
    'xlabel':'Stages',
    'ylabel':'Scores',
}
xs = [1,2,3,4,5]
ys = [1,2,3,4,5]

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
# ax.set(title='My first matplotlib plot', xlabel='stages', ylabel='scores')
ax.set(**props)
ax.plot(xs, ys)
```

```python
# 1.1.4.1 범례 추가하기
# 범례를 추가하는 가장 쉬운 방법은 각 그래프에 label 인자를 넘기는 것임
# 그래프에는 3개의 데이터 곡선이 있고 label이 one, two, three로 지정한다

from numpy.random import randn
fig = plt.figure(); ax = fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(), 'k', label='one')
ax.plot(randn(1000).cumsum(), 'k--', label='two')
ax.plot(randn(1000).cumsum(), 'k.', label='three')
```

```python
#범례를 표시하도록 설정한다. 위치는 best(범례가 표시되기 좋은 곳을 자동으로 설정)
ax.legend(loc='best')
fig
```

```python
# 1.1.5 주석과 그림 추가하기
# 주석과 글자는 text, arrow, annotate 함수를 이용해서 추가할 수 있다.
# text 함수는 그래프 내의 주어진 좌표 (x,y)에 부가적인 스타일로 글자를 그려준다
# ax.text(x, y, 'Hello world!', family='monospace', fontsize=10)
import os
import urllib.request

#파일 다운로드
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/kumi1688/data_repo/main/"
datapath = os.path.join("")
filename = 'spx.csv'
print("Downloading", filename)
url = DOWNLOAD_ROOT + filename
urllib.request.urlretrieve(url, datapath + filename)
```

```python
#아래의 그래프는 야후! 파이낸스에서 얻은 2007년부터의 S&P 500 지수 데이터로 그래프를 생성하고
#2008-2009년 사이에 있었던 재정위기 중 중요한 날짜를 주석으로 추가한 것이다
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# 데이터의 경로를 명시해준다
path = 'spx.csv'
# 데이터를 드라이브에서 읽은 다음
data = pd.read_csv(path, index_col = 0, parse_dates=True)
# 데이터에서 SPX(S&P 500)에 관련된 부분만 추출한다
spx = data['SPX']

#SPX를 그린다
spx.plot(ax=ax, style='k-')

#위기가 있었던 시점을 표시한다
#앞은 날짜, 뒤는 해당 날짜에 찾아온 위기의 이름
crisis_data = [
    (datetime(2007,10,11), 'Peak of bull market'),
    (datetime(2008,3,12), 'Bear Stearns Fails'),
    (datetime(2008,9,15), 'Lehman Bankruptcy'),
]

#위기가 있었던 시점을 직접 그래프에 표시할 방법을 선택
#위기가 있었던 날짜와 위기의 이름을 받아옴
#위기의 이름을 표시하되, 위기가 있었던 날짜의 75 픽셀 또는 225 픽셀만큼 올려서 표현
#화살표를 그리되 검은색과 화살표의 크기를 지정하는 옵션을 준다
#위기의 이름, 화살표를 가로로 왼쪽, 세로로 상단에 위치하도록 정렬
for date, label in crisis_data:
  ax.annotate(label,
      xy=(date, spx.asof(date) + 75),
      xytext=(date, spx.asof(date) + 225),
      arrowprops=dict(facecolor='black', headwidth=4, width=2,
              headlength=4),
      horizontalalignment='left', verticalalignment='top'
      )

#날짜를 2007년에서 2011년 사이로 한정한다
#주가는 600이상 1800미만이므로 이 범위로 한정한다
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
#제목을 표시한다
ax.set_title('Important dates in the 2008-2009 financial crisis')
```
